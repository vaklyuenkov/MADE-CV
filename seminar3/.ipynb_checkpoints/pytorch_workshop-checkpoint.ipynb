## 2. sklearn logreg

```python
from sklearn.datasets import make_blobs
import sklearn
X,y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=((10,5),(5,-5)),
    cluster_std=3,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=0,
)

plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
```

```python

w1 = model.coef_[0][0]
w2 = model.coef_[0][1]
b = model.intercept_[0]

def plot_decision_line(w1,w2,b):
    plt.figure(figsize = (20,10))
    plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
    plt.legend(['y= -1','y= 1'])
    x_arr = np.linspace(-1,18,1000)
    y_arr = (w1*x_arr+b)/w2
    plt.grid(True)
    plt.plot(x_arr,y_arr, linewidth=4)
    plt.show()

plot_decision_line(w1,w2,b)
```

## 3. torch logreg

```python
from sklearn.datasets import make_blobs
import sklearn
X,y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=((10,5),(5,-5)),
    cluster_std=3,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=0,
)
```

```python
x = torch.tensor(X)
y = torch.tensor(y, dtype =torch.double)

def BCE(y,preds):
        #pred = torch.cmax(pred, 1e-5)
        #(1/y.shape[0])
        return -1*(y*torch.log(preds) + (1-y)*torch.log(1-preds))
```

```python

alpha = 0.001
w = torch.ones(2,requires_grad=True, dtype =torch.double )
b = torch.ones(1,requires_grad=True, dtype =torch.double )

for i in range(100):
    clear_output(wait=True)
    preds = torch.sigmoid(torch.matmul(x,w) +b)
    loss = torch.sum(BCE(y,preds))
    loss.backward()

    w.data -= alpha*w.grad.data
    b.data -= alpha*b.grad.data
    
    w.grad.zero_()
    b.grad.zero_()
    
    plot_decision_line(w.data[0] ,w.data[1] ,b.data[0] ) 
    print(f'iter:{i} loss:{float(loss)}')
```

## 4. torch logreg with torch implemented functions

```python
from sklearn.datasets import make_blobs
import sklearn
X,y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=((10,5),(5,-5)),
    cluster_std=3,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=0,
)
```

```python
x = torch.tensor(X)
y = torch.tensor(y, dtype =torch.double)

model = torch.nn.Linear(2,1)
criterion = torch.nn.BCELoss()
optim = torch.optim.SGD(model.parameters(),lr=0.01)
model.train()

for i in range(100):
    clear_output(wait=True)
    
    preds = torch.sigmoid(model(x.float()))
    loss = criterion(preds, y.float())
    loss.backward()
    optim.step()
    optim.zero_grad()
    
    plot_decision_line(model.weight.data[0][0] ,model.weight.data[0][1], model.bias.data[0] ) 
    print(f'iter:{i} loss:{float(loss)}')
```

## 5. non-linear task

```python
from sklearn.datasets import make_moons
X,y = make_moons(
    n_samples=200,
    noise=.1,
    random_state=17,
)

plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
```

```python
def plot_decision_line(w1,w2,b):
    plt.figure(figsize = (20,10))
    plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
    plt.legend(['y= -1','y= 1'])
    x_arr = np.linspace(-1,2,1000)
    y_arr = (w1*x_arr+b)/w2
    plt.grid(True)
    plt.plot(x_arr,y_arr, linewidth=4)
    plt.show()
```

```python
def plot_decision_space(model):
    h = 0.02
    cm = plt.cm.RdBu
    x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
    y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    
    
    input_tensor = torch.from_numpy(np.c_[xx.ravel(),yy.ravel()]).float()
    #model.eval()
    Z = torch.sigmoid(model(input_tensor))
    Z = Z.reshape(xx.shape)
    
    plt.figure(figsize = (20,10))
    plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
    plt.legend(['y= -1','y= 1'])
    plt.contourf(xx,yy,Z.detach().numpy(), cmap =cm, alpha = 0.8)
    plt.show()
```

```python
x = torch.tensor(X)
y = torch.tensor(y, dtype =torch.double)

model = torch.nn.Sequential()
model.add_module("first_layer",nn.Linear(2,2))
model.add_module("first_activation",nn.Sigmoid())
model.add_module("last_layer",nn.Linear(2,1))
criterion = torch.nn.BCELoss()
optim = torch.optim.SGD(model.parameters(),lr=3)
model.train()

for i in range(30000):
    clear_output(wait=True)
    
    preds = torch.sigmoid(model(x.float()))
    loss = criterion(preds, y.float())
    loss.backward()
    optim.step()
    optim.zero_grad()
    if (i+1)%500==0:
        plot_decision_space(model) 
        print(f'iter:{i} loss:{float(loss)}')
```

## 6. non-linear task, better model

```python
x = torch.tensor(X)
y = torch.tensor(y, dtype =torch.double)

model = torch.nn.Sequential()
model.add_module("first_layer",nn.Linear(2,5))
model.add_module("first_activation",nn.Sigmoid())
model.add_module("second_layer",nn.Linear(5,2))
model.add_module("second_activation",nn.Sigmoid())
model.add_module("output_layer",nn.Linear(2,1))
criterion = torch.nn.BCELoss()
optim = torch.optim.SGD(model.parameters(),lr=3)
model.train()

for i in range(30000):
    clear_output(wait=True)
    
    preds = torch.sigmoid(model(x.float()))
    loss = criterion(preds, y.float())
    loss.backward()
    optim.step()
    optim.zero_grad()
    if (i+1)%500==0:
        plot_decision_space(model) 
        print(f'iter:{i} loss:{float(loss)}')
```

## 7. More than 2 classes

```python
from sklearn.datasets import make_blobs
import sklearn
X,y = make_blobs(
    n_samples=400,
    n_features=2,
    cluster_std=[1.0,1.5,0.5],
    #centers=((10,5),(5,-5)),
    #cluster_std=3,
    #center_box=(-10.0, 10.0),
    #shuffle=True,
    random_state=17,
)
plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
```

```python
def plot_decision_space(model):
    h = 0.02
    cm = plt.cm.RdBu
    x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
    y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    
    
    input_tensor = torch.from_numpy(np.c_[xx.ravel(),yy.ravel()]).float()

    Z = model(input_tensor)
    Z = torch.argmax(Z, axis=1)
    Z = Z.reshape(xx.shape)
    
    plt.figure(figsize = (20,10))
    plt.scatter(X[:,0],X[:,1], marker='o', c=y,s=25,edgecolor = 'k')
    plt.contourf(xx,yy,Z.detach().numpy(), cmap =cm, alpha = 0.8)
    plt.show()
```

```python
x = torch.tensor(X)
y = torch.tensor(y)

model = torch.nn.Sequential()
model.add_module("first_layer",nn.Linear(2,5))
model.add_module("first_activation",nn.Sigmoid())
model.add_module("second_layer",nn.Linear(5,5))
model.add_module("second_activation",nn.Sigmoid())
model.add_module("output_layer",nn.Linear(5,3))
model.add_module("output_activation",nn.Sigmoid())
criterion = torch.nn.CrossEntropyLoss()
optim = torch.optim.SGD(model.parameters(),lr=0.1)
model.train()

for i in range(30000):
    clear_output(wait=True)
    
    preds = model(x.float())
    loss = criterion(preds, y)
    loss.backward()
    optim.step()
    optim.zero_grad()
    if (i+1)%500==0:
        plot_decision_space(model) 
        print(f'iter:{i} loss:{float(loss)}')
```
